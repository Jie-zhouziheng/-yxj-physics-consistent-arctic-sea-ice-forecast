```python
import xarray as xr, glob
files = sorted(glob.glob("../data/raw/nsidc_sic/NSIDC0051_SEAICE_PS_N25km_2000*.nc"))
datasets = [xr.open_dataset(f, engine="netcdf4") for f in files]
ds = xr.concat(datasets, dim="time", combine_attrs="drop_conflicts")
print(ds)
print(ds.data_vars)
# 用完可以关闭：for d in datasets: d.close()
```
